Transaction count per day of the week

```python
df['DayofWeek'] = df['Date and Time'].dt.day_name()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_transaction_count_by_dayofweek = df.groupby('DayofWeek')['Description'].count().reindex(day_order)

plt.figure(figsize=(10, 6))
sns.barplot(x=daily_transaction_count_by_dayofweek.index, y=daily_transaction_count_by_dayofweek.values, color='lightgreen')
plt.title('Transaction Count per Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Transaction Count')
plt.show()
```

```python
# Assuming df is your DataFrame with columns 'Date and Time' and 'Description'

df['DayofWeek'] = df['Date and Time'].dt.day_name()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_transaction_count_by_dayofweek = df.groupby('DayofWeek')['Description'].count().reindex(day_order)

# Calculate percent increase
percent_increase = daily_transaction_count_by_dayofweek.pct_change() * 100

plt.figure(figsize=(10, 6))

# Plot the bar chart
ax1 = sns.barplot(x=daily_transaction_count_by_dayofweek.index, y=daily_transaction_count_by_dayofweek.values, color='lightgreen')
plt.title('Transaction Count per Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Transaction Count')

# Create a second y-axis to show percent increase
ax2 = ax1.twinx()
ax2.plot(daily_transaction_count_by_dayofweek.index, percent_increase, color='blue', marker='o', label='Percent Increase')

# Set y-axis label for the second y-axis
ax2.set_ylabel('Percent Increase', color='blue')

# Show legend for the second y-axis
ax2.legend(loc='upper right')

plt.show()
```

Hourly transaction count

```python
df['Hour'] = df['Date and Time'].dt.hour
hourly_transaction_count = df.groupby('Hour')['Description'].count()

plt.figure(figsize=(10, 5))
sns.lineplot(x=hourly_transaction_count.index, y=hourly_transaction_count.values, marker='o', color='orange')
plt.title('Hourly Transaction Count')
plt.xlabel('Hour of the Day')
plt.ylabel('Transaction Count')
plt.show()
```

```python
df['Hour'] = df['Date and Time'].dt.hour
hourly_transaction_count = df.groupby('Hour')['Description'].count()

plt.figure(figsize=(10, 5))
sns.lineplot(x=hourly_transaction_count.index, y=hourly_transaction_count.values, marker='o', color='orange')
tick_positions = range(1, 25)  # 1 to 24 for 24 hours
xtick_labels = [f"{(hour - 1) % 12 + 1}{'am' if (hour - 1) < 12 else 'pm'}" for hour in xtick_positions]

plt.xticks(xtick_positions, xtick_labels, rotation=45, ha='right')

plt.title('Hourly Transaction Count')
plt.xlabel('Hour of the Day')
plt.ylabel('Transaction Count')

# Highlight five peaks with enlarged markers
peaks = hourly_transaction_count.nlargest(5).index
peak_values = hourly_transaction_count.nlargest(5).values

plt.scatter(peaks, peak_values, color='red', marker='^', s=200, label='Peaks')  # Enlarge the marker size (s=100)
plt.show()
```

```python
# Extract the Hour from 'Date and Time'
df['Hour'] = df['Date and Time'].dt.hour

# Count transactions by hour
hourly_transaction_count = df.groupby('Hour')['Description'].count()

plt.figure(figsize=(10, 5))
sns.lineplot(x=hourly_transaction_count.index, y=hourly_transaction_count.values, marker='o', color='orange')

# Set x-axis ticks and labels
xtick_positions = range(1, 25)  # 1 to 24 for 24 hours
xtick_labels = [f"{(hour - 1) % 12 + 1}{'pm' if (hour - 1) < 12 else 'am' if hour != 12 else 'pm'}" for hour in xtick_positions]

plt.xticks(xtick_positions, xtick_labels, rotation=45, ha='right')

plt.title('Hourly Transaction Count')
plt.xlabel('Hour of the Day')
plt.ylabel('Transaction Count')

# Highlight five peaks with enlarged markers
peaks = hourly_transaction_count.nlargest(5).index
peak_values = hourly_transaction_count.nlargest(5).values

plt.scatter(peaks, peak_values, color='red', marker='^', s=100, label='Peaks')  # Enlarge the marker size (s=100)

# Show legend
plt.legend()

plt.show()
```
